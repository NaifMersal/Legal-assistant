```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import SentenceTransformerEmbeddings

# Load the same embedding model
embedding_model = SentenceTransformerEmbeddings(model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

# Reconnect to your existing ChromaDB
vectorstore = Chroma(
    persist_directory="chroma_db",
    embedding_function=embedding_model
)

# Count stored documents
count = vectorstore._collection.count()
print(f"Total stored documents: {count}")

# Test: Retrieve top 3 articles for a sample query
query = "ما هو النظام الأساسي للحكم في المملكة العربية السعودية؟"
results = vectorstore.similarity_search(query, k=3)

print("\nTop 3 search results:\n")
for i, doc in enumerate(results, 1):
    print(f"{i}. {doc.metadata['law_name']} — {doc.metadata['article_title']}")
    print(doc.page_content[:200], "...\n")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# Same embedding model used for Chroma
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

# Load your persisted database
chroma_db = Chroma(persist_directory="chroma_db", embedding_function=embedding_model)
retriever = chroma_db.as_retriever(search_kwargs={"k": 5})
```

```python
# Create Gemini chat model
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",  # Fast and cheap; can change to gemini-2.5-pro for deeper reasoning
    temperature=0.2
)

# Define a custom Arabic-friendly prompt
prompt_template = """
السؤال:
{question}

المواد ذات الصلة:
{context}

بناءً على المواد السابقة، قدم إجابة دقيقة ومبسطة للسؤال أعلاه. 
إذا لم تجد الإجابة في النصوص، قل: "المعلومات غير متوفرة في النظام."
"""

prompt = PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)

# Build the RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",  # simplest chain type: stuff retrieved docs into prompt
    retriever=retriever,
    chain_type_kwargs={"prompt": prompt}
)
```

```python
query = "ما هي الحالات التي يجوز فيها إنهاء عقد العمل بدون إشعار؟"
result = qa_chain.run(query)
print(result)
```

```python
docs = retriever.get_relevant_documents(query)
for i, doc in enumerate(docs[:3]):
    print(f"\n--- مادة رقم {i+1} ---\n{doc.page_content[:300]}...")
```
