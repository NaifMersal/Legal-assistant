```python
from sentence_transformers import SentenceTransformer

# Load the multilingual model
model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

# Example Arabic texts
texts = [
    "نظام العمل السعودي ينظم العلاقة بين العامل وصاحب العمل.",
    "القانون السعودي يحمي حقوق العاملين في القطاع الخاص."
]

# Generate embeddings
embeddings = model.encode(texts)

print("Embedding shape:", embeddings.shape)
print("Sample embedding (first vector):", embeddings[0][:10])  # print first 10 dims
```

```python
import json

# Load the data
with open("../laws.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Recursive function to count all article dictionaries
def count_articles(obj):
    count = 0
    if isinstance(obj, dict):
        for v in obj.values():
            count += count_articles(v)
    elif isinstance(obj, list):
        # Each element here is likely a dictionary (Article_Title/Text)
        count += len(obj)
    return count

# Count total
total_articles = count_articles(data)
print(f"Total number of articles found: {total_articles}")
```
